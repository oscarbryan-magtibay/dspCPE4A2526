```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('coins.jpeg')
image = cv2.resize(image, (640, 480))  # Resize if needed

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

lower_silver = np.array([0, 0, 120])
upper_silver = np.array([180, 60, 255])

lower_gold = np.array([15, 80, 100])
upper_gold = np.array([35, 255, 255])

silver_mask = cv2.inRange(hsv, lower_silver, upper_silver)
gold_mask = cv2.inRange(hsv, lower_gold, upper_gold)

silver_result = cv2.bitwise_and(image, image, mask=silver_mask)
gold_result = cv2.bitwise_and(image, image, mask=gold_mask)

fig, axs = plt.subplots(1, 3, figsize=(18, 12))
axs[0].imshow(image_rgb)
axs[0].set_title("Original")
axs[0].axis("off")

axs[1].imshow(silver_mask, cmap="gray")
axs[1].set_title("Silver Mask")
axs[1].axis("off")

axs[2].imshow(gold_mask, cmap="gray")
axs[2].set_title("Gold Mask")
axs[2].axis("off")

plt.tight_layout()
plt.show()
```
