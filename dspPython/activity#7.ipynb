```python
import numpy as np
import matplotlib.pyplot as plt

# Given sequence
xn = [5, -1, 8, 12, 0, -4, 9]

# Ensure length is odd to identify origin
if len(xn) % 2 == 1:
    N = len(xn)
    n = np.arange(-(N//2), N//2 + 1)  
    x = np.array(xn)

    # Time-reversed, Even, and Odd 
    x_rev = x[::-1]
    x_even = 0.5 * (x + x_rev)  
    x_odd = 0.5 * (x - x_rev)   

    print("Original sequence:", x)
    print("Time-reversed:    ", x_rev)
    print("Even:   ", x_even)
    print("Odd:    ", x_odd)

    # Plotting
    plt.figure(figsize=(10,8))

    plt.subplot(4,1,1)
    plt.stem(n, x)
    plt.title("Original Sequence (x[n])")
    plt.xlabel("n"); plt.ylabel("x[n]")

    plt.subplot(4,1,2)
    plt.stem(n, x_rev)
    plt.title("Time-Reversed (x[-n])")
    plt.xlabel("n"); plt.ylabel("x[-n]")

    plt.subplot(4,1,3)
    plt.stem(n, x_even)
    plt.title("Even (x_e[n])")
    plt.xlabel("n"); plt.ylabel("x_e[n]")

    plt.subplot(4,1,4)
    plt.stem(n, x_odd)
    plt.title("Odd (x_o[n])")
    plt.xlabel("n"); plt.ylabel("x_o[n]")

    plt.tight_layout()
    plt.show()

else:
   print("Sequence length is even — cannot determine origin")
```
