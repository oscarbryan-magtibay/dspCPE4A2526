```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5,10)
x = (2.0 ** n)
print("n =", n)
print("x =", x)

```

```python
plt.stem(n, x)
plt.title('Sequence x[n] = 2^n')
plt.xlabel('n (sample index)')
plt.ylabel('x[n]')
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3.14, 3.14 + 0.01, 0.01)
y = np.sin(x)

plt.figure()
markerline, stemlines, baseline = plt.stem(x, y)
plt.title("Sinusoidal Sequence")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 50, 1)
y = np.log(x)

plt.figure()
plt.stem(x, y)
plt.title("Natural Logarithm x[n] = ln(n)")
plt.xlabel("x")
plt.ylabel("ln(x)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
