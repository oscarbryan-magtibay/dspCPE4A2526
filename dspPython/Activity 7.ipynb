```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence and time indices
xn = [0, -1, 0, 3, 0]
n_values = np.arange(-(len(xn)//2), len(xn)//2 + 1)
x_signal = np.array(xn)

# Check if sequence length is odd (has a defined origin)
if len(x_signal) % 2 == 0:
    print("Error: Sequence length must be odd to properly define the origin")
else:
    # Calculate even and odd components
    x_reversed = x_signal[::-1]
    even_component = 0.5 * (x_signal + x_reversed)
    odd_component = 0.5 * (x_signal - x_reversed)
    
    # Create the plot
    fig, axes = plt.subplots(3, 1, figsize=(10, 8))
    
    # Plot original signal
    axes[0].stem(n_values, x_signal, linefmt='b-', markerfmt='bo', basefmt='r-')
    axes[0].set_title('Original Signal: x[n]')
    axes[0].set_xlabel('Sample Index (n)')
    axes[0].set_ylabel('Amplitude')
    axes[0].grid(True, alpha=0.3)
    axes[0].axhline(0, color='red', linewidth=0.8)
    axes[0].axvline(0, color='red', linewidth=0.8)
    
    # Plot even component
    axes[1].stem(n_values, even_component, linefmt='g-', markerfmt='go', basefmt='r-')
    axes[1].set_title('Even Component: xₑ[n]')
    axes[1].set_xlabel('Sample Index (n)')
    axes[1].set_ylabel('Amplitude')
    axes[1].grid(True, alpha=0.3)
    axes[1].axhline(0, color='blue', linewidth=0.8)
    axes[1].axvline(0, color='blue', linewidth=0.8)
    
    # Plot odd component
    axes[2].stem(n_values, odd_component, linefmt='m-', markerfmt='mo', basefmt='r-')
    axes[2].set_title('Odd Component: xₒ[n]')
    axes[2].set_xlabel('Sample Index (n)')
    axes[2].set_ylabel('Amplitude')
    axes[2].grid(True, alpha=0.3)
    axes[2].axhline(0, color='green', linewidth=0.8)
    axes[2].axvline(0, color='green', linewidth=0.8)
    
    # Adjust layout and display
    plt.tight_layout()
    plt.show()
    
    # Print some information about the decomposition
    print(f"Original signal: {x_signal}")
    print(f"Even component: {even_component}")
    print(f"Odd component: {odd_component}")
    print(f"Verification (x_e + x_o): {even_component + odd_component}")
```
