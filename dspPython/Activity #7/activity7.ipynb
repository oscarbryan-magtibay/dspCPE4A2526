```python
#Original Signal

import numpy as np
import matplotlib.pyplot as plt

xn = np.array([1, -3, 5, -7, 9])
N = (len(xn))

n0 = N//2
n = np.arange(-n0, N - n0)

plt.stem(n, xn)
plt.title("Original Signal x[n]")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.show()
```

```python
#Folded Signal

import numpy as np
import matplotlib.pyplot as plt

xn = np.array([1, -3, 5, -7, 9])
N = (len(xn))

n0 = N//2
n = np.arange(-n0, N - n0)

x_fold = xn[::-1]

plt.stem(n, x_fold)
plt.title("Folded Signal x[-n]")
plt.xlabel("n")
plt.ylabel("x[-n]")
plt.grid(True)
plt.show()
```

```python
#Even Component

import numpy as np
import matplotlib.pyplot as plt

xn = np.array([1, -3, 5, -7, 9])
N = (len(xn))

n0 = N//2
n = np.arange(-n0, N - n0)

x_fold = xn[::-1]
x_even = 0.5 * (xn + x_fold)

plt.stem(n, x_even)
plt.title("Even Component")
plt.xlabel("n")
plt.ylabel("x_even[n]")
plt.grid(True)
plt.show()
```

```python
#Odd Component

import numpy as np
import matplotlib.pyplot as plt

xn = np.array([1, -3, 5, -7, 9])
N = (len(xn))

n0 = N//2
n = np.arange(-n0, N - n0)

x_fold = xn[::-1]
x_odd = 0.5 * (xn - x_fold)

plt.stem(n, x_odd)
plt.title("Odd Component")
plt.xlabel("n")
plt.ylabel("x_odd[n]")
plt.grid(True)
plt.show()
```
